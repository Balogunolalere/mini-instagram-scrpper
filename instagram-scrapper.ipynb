```python
flist = driver.find_element(By.CLASS_NAME, 'isgrP')
scroll = 0

while scroll < 15:
    driver.execute_script('arguments[0].scrollTop = arguments[0].scrollTop + arguments[0].offsetHeight;', flist)
    time.sleep(3)
    scroll += 1
    f_list = driver.find_elements(By.TAG_NAME, 'li')
    if len(f_list) == n_follower:
        break

for f in f_list:
    wk1 = f.text.replace('Follow', ' ')
    print(wk1.replace('Remove', ' ') + '\n')
```
